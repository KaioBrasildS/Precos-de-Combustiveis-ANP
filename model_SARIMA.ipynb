### Modelo SARIMA(1,2,1)(1,2,1)[12]

Esse modelo inclui componentes autorregressivos e de média móvel tanto para as partes não sazonais quanto sazonais, além de dois níveis de diferenciação para ambas as partes.

### Fórmula Compacta do Modelo SARIMA(1,2,1)(1,2,1)[12]

A forma compacta do modelo SARIMA(1,2,1)(1,2,1)[12] pode ser escrita como:

$$
(1 - \Phi_1 B^{12}) (1 - \phi_1 B)(1 - B)^2(1 - B^{12})^2 Y_t = (1 + \Theta_1 B^{12})(1 + \theta_1 B) \epsilon_t
$$

Onde:
- $Y_t$ é o valor da série temporal no tempo $t$.
- $B$ é o operador de defasagem (lag operator), onde $B^k Y_t = Y_{t-k}$.
- $\phi_1(B)$ é o polinômio autorregressivo não sazonal de ordem $1$.
- $\Phi_1(B^s)$ é o polinômio autorregressivo sazonal de ordem $1$.
- $\theta_1(B)$ é o polinômio de média móvel não sazonal de ordem $2$.
- $\Theta_1(B^12)$ é o polinômio de média móvel sazonal de ordem $2$.
- $\epsilon_t$ é o termo de erro (resíduo) no tempo $t$.
- $(1 - B)^2$ representa a diferenciação não sazonal de ordem $2$.
- $(1 - B^{12})^2$ representa a diferenciação sazonal de ordem $2$.
### Fórmula Expandida do Modelo SARIMA(1,2,1)(1,2,1)[12]

Expandindo as partes autorregressiva (AR) e de média móvel (MA):

1. **Parte Autorregressiva (AR):**

 Expansão da Parte Autorregressiva (AR)

A parte AR do modelo SARIMA(1,2,1)(1,2,1)[12] pode ser expressa como:

$ (1 - \phi_1 B)(1 - \Phi_1 B^{12})(1 - B)^2(1 - B^{12})^2 Y_t $

Podemos expandir e simplificar essa equação passo a passo:

* Expansão do Componente AR Não Sazonal

   * O termo 


   $$ (1 - \phi_1 B) Y_t = Y_t - \phi_1 Y_{t-1} $$


* Expansão do Componente AR Sazonal

   * O termo 


   $$ (1 - \Phi_1 B^{12}) Y_t = Y_t - \Phi_1 Y_{t-12} $$


* Multiplicação Cruzada dos Termos AR Não Sazonais e Sazonais

   * Quando multiplicamos os termos AR não sazonais e sazonais, obtemos:


$$ (1 - \phi_1 B)(1 - \Phi_1 B^{12}) Y_t = Y_t - \phi_1 Y_{t-1} - \Phi_1 Y_{t-12} + \phi_1 \Phi_1 Y_{t-13} $$


Esse resultado mostra que o valor de $ Y_t $ depende de seus valores defasados de 1, 12, e 13 períodos, multiplicados pelos coeficientes AR.

2. **Parte de Média Móvel (MA):**

$$
(1 + \theta_1 B)(1 + \Theta_1 B^{12}) \epsilon_t
$$

### Forma Expandida Final

Combinação das partes AR e MA:

$$
Y_t = \phi_1 Y_{t-1} + \Phi_1 Y_{t-12} - \phi_1 \Phi_1 Y_{t-13} + (1 + \theta_1 B)(1 + \Theta_1 B^{12}) \epsilon_t
$$

A forma expandida mostra como os valores passados da série e os erros passados afetam o valor atual.

### Separação da Fórmula SARIMA em Parte AR e Parte MA

#### Fórmula Original

A fórmula original do modelo SARIMA(1,2,1)(1,2,1)[12], antes de ser expandida, é:

$$
\Phi(B^{12}) \phi(B) (1 - B)^2 (1 - B^{12})^2 Y_t = \Theta(B^{12}) \theta(B) \epsilon_t
$$

Onde:
- $\Phi(B^{12}) = (1 - \Phi_1 B^{12})$
- $\phi(B) = (1 - \phi_1 B)$
- $\Theta(B^{12}) = (1 + \Theta_1 B^{12})$
- $\theta(B) = (1 + \theta_1 B)$
- $\epsilon_t$ é o termo de erro.

#### Parte AR (Autorregressiva)

A parte AR da equação é:

$$
(1 - \phi_1 B)(1 - \Phi_1 B^{12})(1 - B)^2 (1 - B^{12})^2 Y_t
$$

- **$(1 - \phi_1 B)$**: Componente autorregressivo não sazonal.
- **$(1 - \Phi_1 B^{12})$**: Componente autorregressivo sazonal.
- **$(1 - B)^2$**: Diferenciação não sazonal de ordem 2.
- **$(1 - B^{12})^2$**: Diferenciação sazonal de ordem 2.

#### Parte MA (Média Móvel)

A parte MA da equação é:

$$
(1 + \theta_1 B)(1 + \Theta_1 B^{12}) \epsilon_t
$$

- **$(1 + \theta_1 B)$**: Componente de média móvel não sazonal.
- **$(1 + \Theta_1 B^{12})$**: Componente de média móvel sazonal.

### Isolando $Y_t$

Para isolar $Y_t$, expandimos a equação AR e movemos a parte MA para o lado direito.

#### Parte AR Expandida

Expandindo a parte AR:

$$
Y_t - \phi_1 Y_{t-1} - \Phi_1 Y_{t-12} + \phi_1 \Phi_1 Y_{t-13} - 2(Y_{t-1} - \phi_1 Y_{t-2} - \Phi_1 Y_{t-13} + \phi_1 \Phi_1 Y_{t-14}) + Y_{t-2} - \phi_1 Y_{t-3} - \Phi_1 Y_{t-14} + \phi_1 \Phi_1 Y_{t-15}
$$

#### Parte MA Expandida

A parte MA expandida é simplesmente aplicada ao erro $\epsilon_t$:

$$
(1 + \theta_1 B)(1 + \Theta_1 B^{12}) \epsilon_t
$$

#### Equação Final Isolando $Y_t$

Com $Y_t$ isolado no lado esquerdo, a equação completa é:

$$
Y_t = \phi_1 Y_{t-1} + \Phi_1 Y_{t-12} - \phi_1 \Phi_1 Y_{t-13} + 2\left( Y_{t-1} - \phi_1 Y_{t-2} - \Phi_1 Y_{t-13} + \phi_1 \Phi_1 Y_{t-14} \right) - Y_{t-2} + \phi_1 Y_{t-3} + \Phi_1 Y_{t-14} - \phi_1 \Phi_1 Y_{t-15} + (1 + \theta_1 B)(1 + \Theta_1 B^{12}) \epsilon_t
$$

Essa equação representa a combinação de todos os componentes autorregressivos, diferenciais, sazonais, e de média móvel que influenciam o valor atual $Y_t$, levando em conta tanto os valores passados da série temporal quanto os erros passados.
