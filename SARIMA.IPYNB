### SARIMA(1,2,1)(1,2,1)[12]

#### Componentes do Modelo SARIMA

1. **Componentes Não Sazonais**:
   - $p = 1$: Ordem do termo autorregressivo (AR) não sazonal.
   - $d = 2$: Ordem de diferenciação não sazonal.
   - $q = 1$: Ordem do termo de média móvel (MA) não sazonal.

2. **Componentes Sazonais**:
   - $P = 1$: Ordem do termo autorregressivo sazonal (AR sazonal).
   - $D = 2$: Ordem de diferenciação sazonal.
   - $Q = 1$: Ordem do termo de média móvel sazonal (MA sazonal).
   - $s = 12$: Período sazonal (12 meses).

#### Fórmula Compacta do Modelo SARIMA(1,2,1)(1,2,1)[12]

A equação compacta do modelo SARIMA(1,2,1)(1,2,1)[12] pode ser escrita como:

$$
\Phi(B^{12}) (1 - B^{12})^2 \phi(B) (1 - B)^2 Y_t = \Theta(B^{12}) \theta(B) \epsilon_t
$$

Onde:
- $\Phi(B^{12}) = (1 - \Phi_1 B^{12})$: Polinômio autorregressivo sazonal.
- $(1 - B^{12})^2$: Diferenciação sazonal de ordem 2.
- $\phi(B) = (1 - \phi_1 B)$: Polinômio autorregressivo não sazonal.
- $(1 - B)^2$: Diferenciação não sazonal de ordem 2.
- $\Theta(B^{12}) = (1 + \Theta_1 B^{12})$: Polinômio de média móvel sazonal.
- $\theta(B) = (1 + \theta_1 B)$: Polinômio de média móvel não sazonal.
- $\epsilon_t$: Termo de erro.

#### Fórmula Expandida

**1. Parte Autorregressiva (AR):**

* Expansão do Componente AR Sazonal e Não Sazonal (sem diferenciação ainda):

$$
(1 - \Phi_1 B^{12})(1 - \phi_1 B) Y_t = Y_t - \phi_1 Y_{t-1} - \Phi_1 Y_{t-12} + \phi_1 \Phi_1 Y_{t-13}
$$

Essa expansão inclui os termos autorregressivos não sazonais (defasados de 1 período) e sazonais (defasados de 12 períodos), além das interações entre eles.

**2. Parte de Média Móvel (MA):**

$$
(1 + \theta_1 B)(1 + \Theta_1 B^{12}) \epsilon_t
$$

Expandindo o polinômio de média móvel:

$$
\epsilon_t + \theta_1 \epsilon_{t-1} + \Theta_1 \epsilon_{t-12} + \theta_1 \Theta_1 \epsilon_{t-13}
$$

**3. Equação sem a Diferenciação:**

Neste ponto, temos a equação completa sem aplicar a diferenciação:

$$
Y_t - \phi_1 Y_{t-1} - \Phi_1 Y_{t-12} + \phi_1 \Phi_1 Y_{t-13} = \epsilon_t + \theta_1 \epsilon_{t-1} + \Theta_1 \epsilon_{t-12} + \theta_1 \Theta_1 \epsilon_{t-13}
$$

#### Aplicando a Diferenciação

Agora, aplicamos a diferenciação sazonal e não sazonal de ordem 2:

1. **Diferenciação Não Sazonal**:

A diferenciação de ordem 2 não sazonal significa que aplicamos $(1 - B)^2$, que é equivalente a:

$$
(1 - 2B + B^2)
$$

Isso resulta em:

$$
Y_t^{(2)} = Y_t - 2Y_{t-1} + Y_{t-2}
$$

Aplicamos essa diferenciação à equação anterior:

$$
(Y_t - 2Y_{t-1} + Y_{t-2}) - \phi_1 (Y_{t-1} - 2Y_{t-2} + Y_{t-3}) - \Phi_1 (Y_{t-12} - 2Y_{t-13} + Y_{t-14}) + \phi_1 \Phi_1 (Y_{t-13} - 2Y_{t-14} + Y_{t-15})
$$

2. **Diferenciação Sazonal**:

A diferenciação sazonal de ordem 2 $(1 - B^{12})^2$ é equivalente a:

$$
(1 - 2B^{12} + B^{24})
$$

Aplicamos essa diferenciação sazonal à equação:

$$
Y_t^{(12)} = Y_t - 2Y_{t-12} + Y_{t-24}
$$

Substituímos na equação expandida:

$$
(Y_t - 2Y_{t-12} + Y_{t-24}) - \phi_1 (Y_{t-1} - 2Y_{t-13} + Y_{t-25}) - \Phi_1 (Y_{t-12} - 2Y_{t-24} + Y_{t-36}) + \phi_1 \Phi_1 (Y_{t-13} - 2Y_{t-25} + Y_{t-37})
$$

#### Equação Final Isolando $Y_t$

A equação final, após aplicar as diferenciações e isolar $Y_t$, é:

$
Y_t = 2Y_{t-1} - Y_{t-2} + \phi_1(2Y_{t-2} - Y_{t-3}) + 2Y_{t-12} - Y_{t-24} + \Phi_1(2Y_{t-13} - Y_{t-25}) + \phi_1\Phi_1(2Y_{t-14} - Y_{t-26}) + \epsilon_t + \theta_1 \epsilon_{t-1} + \Theta_1 \epsilon_{t-12} + \theta_1 \Theta_1 \epsilon_{t-13}
$

Essa equação final reflete a combinação dos efeitos autorregressivos e de média móvel, sazonais e não sazonais, junto com as diferenciações de ordem 2 aplicadas posteriormente.


* Aplicar primeiro as diferenças sazonais dentro do parentesese após isso aplicar a diferenciação não sazonal.

## Modelo SARIMA(1,2,1)(1,2,1)[12]

### Componentes do Modelo SARIMA

#### Componentes Não Sazonais:

- **\(p = 1\)**: Ordem do termo autorregressivo (AR) não sazonal.
- **\(d = 2\)**: Ordem de diferenciação não sazonal.
- **\(q = 1\)**: Ordem do termo de média móvel (MA) não sazonal.

#### Componentes Sazonais:

- **$ P = 1$**: Ordem do termo autorregressivo sazonal (AR sazonal).
- **$ D = 2 $**: Ordem de diferenciação sazonal.
- **$ Q = 1 $**: Ordem do termo de média móvel sazonal (MA sazonal).
- **$ s = 12 $**: Período sazonal (12 meses).

### Equação sem a Diferenciação:

Neste ponto, temos a equação completa sem aplicar a diferenciação:

$$
Y_t - \phi_1 Y_{t-1} - \Phi_1 Y_{t-12} + \phi_1 \Phi_1 Y_{t-13} = \epsilon_t + \theta_1 \epsilon_{t-1} + \Theta_1 \epsilon_{t-12} + \theta_1 \Theta_1 \epsilon_{t-13}
$$


### Fórmula SARIMA com  $ W_t $

A fórmula compacta do modelo SARIMA(1,2,1)(1,2,1)[12], aplicando a diferenciação sazonal e não sazonal representada por $ W_t = \nabla^d(\nabla_m^D Y_t) $, é:

$$
W_t - \phi_1 W_{t-1} - \Phi_1 W_{t-12} + \phi_1 \Phi_1 W_{t-13} = \epsilon_t + \theta_1 \epsilon_{t-1} + \Theta_1 \epsilon_{t-12} + \theta_1 \Theta_1 \epsilon_{t-13}
$$

Onde:

- $ W_t = \nabla^d(\nabla_m^D Y_t) $ é a série temporal diferenciada.
- $ \phi_1 $ é o coeficiente do termo AR não sazonal.
- $ \Phi_1 $ é o coeficiente do termo AR sazonal.
- $ \theta_1 $ é o coeficiente do termo MA não sazonal.
- $ \Theta_1 $ é o coeficiente do termo MA sazonal.
- $ \epsilon_t $ é o termo de erro branco.

### Parte Autorregressiva (AR):

- **Parte Autorregressiva Não Sazonal (AR):**

$$
(1 - \phi_1 B)W_t = W_t - \phi_1 W_{t-1}
$$

- **Parte Autorregressiva Sazonal (AR Sazonal):**

$$
(1 - \Phi_1 B^{12})W_t = W_t - \Phi_1 W_{t-12}
$$

### Parte de Média Móvel (MA):

- **Parte de Média Móvel Não Sazonal (MA):**

$$
(1 + \theta_1 B)\epsilon_t = \epsilon_t + \theta_1 \epsilon_{t-1}
$$

- **Parte de Média Móvel Sazonal (MA Sazonal):**

$$
(1 + \Theta_1 B^{12})\epsilon_t = \epsilon_t + \Theta_1 \epsilon_{t-12}
$$

### Fórmula Expandida com $ W_t $ Isolado

Aplicando a diferenciação e isolando $ W_t $:

$$
W_t = \epsilon_t + \theta_1 \epsilon_{t-1} + \Theta_1 \epsilon_{t-12} + \theta_1 \Theta_1 \epsilon_{t-13} + \phi_1 W_{t-1} + \Phi_1 W_{t-12} - \phi_1 \Phi_1 W_{t-13}
$$


$$
W_t = \nabla^d (\nabla_m^D Y_t)
$$

1. **Diferenciação Sazonal** $\nabla_m^D Y_t$:
   A diferenciação sazonal de ordem $D$ e período $m$ é aplicada primeiro:

   $$
   \nabla_m^D Y_t = (1 - B^m)^D Y_t
   $$

   Para $D = 2$ e $m = 12$, temos:

   $$
   \nabla_m^2 Y_t = (1 - B^{12})^2 Y_t = Y_t - 2Y_{t-12} + Y_{t-24}
   $$

2. **Diferenciação Não Sazonal** $\nabla^d$:
   Agora aplicamos a diferenciação não sazonal de ordem $d$:

   $$
   \nabla^d (Y_t) = (1 - B)^d Y_t
   $$

   Para $d = 2$, temos:

   $$
   \nabla^2 Y_t = (1 - B)^2 Y_t = Y_t - 2Y_{t-1} + Y_{t-2}
   $$

3. **Aplicando a diferenciação não sazonal no resultado da diferenciação sazonal**:

   Aplicamos $\nabla^2$ em $\nabla_m^2 Y_t$:

   $$
   W_t = \nabla^2 (Y_t - 2Y_{t-12} + Y_{t-24})
   $$

   Isso resulta em:

   $$
   W_t = (Y_t - 2Y_{t-1} + Y_{t-2}) - 2(Y_{t-12} - 2Y_{t-13} + Y_{t-14}) + (Y_{t-24} - 2Y_{t-25} + Y_{t-26})
   $$

Portanto, a fórmula expandida de $W_t = \nabla^d (\nabla_m^D Y_t)$ para $d = 2$ e $D = 2$ é:

$$
W_t = Y_t - 2Y_{t-1} + Y_{t-2} - 2(Y_{t-12} - 2Y_{t-13} + Y_{t-14}) + (Y_{t-24} - 2Y_{t-25} + Y_{t-26})
$$
